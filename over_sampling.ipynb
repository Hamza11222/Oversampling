```python
data = pd.read_csv('hiring1 (1).csv')
data.head(100)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(data['EDUCATION'])
education=le.classes_
data['EDUCATION']=le.transform(data['EDUCATION'])
data
```

```python
data.isnull().sum()
```

```python
data['Salary'].fillna(value=data['Salary'].mean(), inplace=True)
```

```python
data['Age'].fillna(value=data['Age'].mean(), inplace=True)
data['EXPEREINCE'].fillna(value=data['EXPEREINCE'].mean(), inplace=True)
```

```python
data['Area'].fillna(value=data['Area'].mode(), inplace=True)
```

```python
data.isnull().sum()
```

```python
targets = data['EDUCATION']
```

```python
data=data.drop(['EDUCATION'], axis = 1)
data
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

le.fit(data['N A M E'])
name=le.classes_
data['N A M E']=le.transform(data['N A M E'])
le.fit(data['Gender'])
gender=le.classes_
data['Gender']=le.transform(data['Gender'])
le.fit(data['Designation'])
designation=le.classes_
data['Designation']=le.transform(data['Designation'])
le.fit(data['Status'])
status=le.classes_
data['Status']=le.transform(data['Status'])
le.fit(data['Joining'])
joining=le.classes_
data['Joining']=le.transform(data['Joining'])
le.fit(data['Area'])
Area=le.classes_
data['Area']=le.transform(data['Area'])
le.fit(data['Department'])
department=le.classes_
data['Department']=le.transform(data['Department'])
le.fit(data['City'])
city=le.classes_
data['City']=le.transform(data['City'])
le.fit(data['Religion'])
religion=le.classes_
data['Religion']=le.transform(data['Religion'])
```

```python

data['Joining'] = data['Joining'].fillna((data['Joining'].max()))
data['Area'] = data['Area'].fillna((data['Area'].max()))
data
```

```python
from imblearn.over_sampling import SMOTE,RandomOverSampler
oversample = RandomOverSampler()
X, y = oversample.fit_resample(data, targets)
```
